```python
def readData(fileName, sheetName):
    # assign to dataframe 
    df = pd.read_excel(fileName, sheet_name=sheetName)

    # replace all white space in cells with nan for consistancy 
    df = df.replace(r'^\s*$', np.nan, regex=True)
    
                
    return(df)
```

```python
# filteredBlends uses the variable blend and returns all datapoints which have that % content
def filterBlends(blend, df):
    
    # if specified use all datapoints 
    if blend == 'All':
        return df

    # else there needs to be a comparison to return the entries of the DF which have the specified HDPE % or more      
    else:
        blend = float(blend)
        df['HDPE content'] = pd.to_numeric(df['HDPE content'],errors='coerce')
        df_mask = df['HDPE content'] >= blend
        blendDf = df[df_mask]

    
        return blendDf
```

```python
def newLinearRegAll(y, x):                                   
    # scikitlearn function 
    LR = LinearRegression()
    
    # training the model with data for the two features being modelled
    LR.fit(x.reshape(-1,1), y)

    # predict the linear regression with data for the two features being modelled
    Prediction = LR.predict(x.reshape(-1,1))
    
    # plotting the reults and the data points 
    plt.plot(x, Prediction, label = 'Linear Regression', c = 'b')
    plt.scatter(x, y, label = 'All Data', c = 'g')
    plt.xlabel(outputProp)
    plt.ylabel(inputProp)
    plt.legend()
    plt.title(outputProp + ' vs ' + inputProp + ' Linear Reg with HDPE perecent:' + str(blend))
    plt.show()
```

```python
# list of inputs and ouputs to fead into the loop for creating the bulk of graphs
# finished product will ask for user input of the input and output feature, given in menu form

listInputs = ['Density', 'Crystallite Size', 'Number Average', 'Weight Average', 'Dispersity','Melting Point', 'Degree of Crystallinity', 'Crystallisation Point']
listOutputs = ['Young\'s Modulus', 'Melt Flow Index', 'Loss/ Storage Modulus', 'Ultimate Tensile Strength','Impact Strength','Elongation at Break','Tensile Strength at Yield']

# list of all blends, after looking at the data I will discount the 'all' moving forward as it creates too many outliers
blends = ['All', 50, 60, 70, 80, 90, 100]
```

```python
global inputProp 
global outputProp 
global blend


# list of features which didnt have enough data points to complete linear regression
notEnough = []

# runs through each HDPE % blend in the list blends
for blend in blends:
    
    # returns the dataframe with just the specified HDPE % content 
    filteredDf = filterBlends(blend, df)
    
    notEnough += [str(blend) + ': ']
    
    # runs through each input property and assigns it to the global var inputProp     
    for i in listInputs:
        inputProp = i
        
        # runs through each output property and assigns it to the global var outputProp    
        for o in listOutputs:
            outputProp = o 
            
            # uses the var outputProp and inputProp to create the smaller DF with the relevant columns             
            colNames = getColNames()
            tempFilteredDf = filteredDf.copy()
            
            # expection handling for function which gets the two np.arrays (inputPropVals, outputPropVals) to feed into the linear regression function 
            try:
                inputPropVals, outputPropVals, tempFilteredDf = getDataFrame(colNames, tempFilteredDf, inputProp, outputProp)
            except Exception as e:
                print("An exception occurred: ", e) 
            
            # expection handling for the linear regression function as some features dont have enough data points so would cause it to crash
            try:
                newLinearRegAll(inputPropVals, outputPropVals)
            except Exception as e:
                print("An exception occurred: ", e)
                
                # if there was an error due to there not being any data points for these features I add it to this list 
                notEnough += [inputProp + ' and ' + outputProp]
```
